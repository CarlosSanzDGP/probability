## Correlation

- **What is correlation**: In statistics, dependence or association is any statistical relationship, whether causal or not, between two random variables or bivariate data. In the broadest sense correlation is any statistical association, though it commonly refers to the degree to which a pair of variables are linearly related.

- **Autocorrelation**: Autocorrelation, also known as serial correlation, is the correlation of a signal with a delayed copy of itself as a function of delay. Informally, it is the `similarity between observations` as a function of the time lag between them.

Correlation Matrix in Pandas `df.corr()`

[https://stackoverflow.com/questions/29432629/plot-correlation-matrix-using-pandas/50703596]

```python
%matplotlib inline
import pandas as pd
import numpy as np

X = np.linspace(0,1,100)
print(X)

Y = pd.DataFrame({
    "Y1": 0.5*X,
    "Y2":-0.5*X+1,
    "Y3": 0.5*X+1,
})
Y.plot()

# Correlation matrix 
display(Y.corr())

# Autocorrelation for Y1
print(Y["Y1"].autocorr(lag=5))
```

## Different correlation metrics

- **Spearman Correlation**: Relationships between variables that are not necessarily linear.
- **Kendall correlation**: Also non linear correlation `df.corr(method="kendall")`


```python
for m in ["pearson", "kendall", "spearman"]:
    print(f"Method: {m}")
    display(Y.corr(method=m))
```
